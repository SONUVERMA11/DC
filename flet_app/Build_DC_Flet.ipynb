# 🎵 DC v2.0 — Build APK with Flet

Builds the DC YouTube MP3 Downloader APK using **Flet** (Flutter-powered).

No Buildozer needed! **Run each cell in order.**

## Step 1: Install Flutter SDK (~2 min)

```
%%bash
# Install Flutter
git clone https://github.com/flutter/flutter.git -b stable --depth 1 /opt/flutter 2>&1 | tail -5
export PATH="/opt/flutter/bin:$PATH"
flutter --version
echo '✅ Flutter installed'
```

## Step 2: Install Flet & Java 17 (~1 min)

```
!pip install -q 'flet>=0.25.0'
!sudo apt-get update -qq && sudo apt-get install -y -qq openjdk-17-jdk
import os
os.environ['JAVA_HOME'] = '/usr/lib/jvm/java-17-openjdk-amd64'
os.environ['PATH'] = '/opt/flutter/bin:' + os.environ['PATH']
print('\n✅ Flet & Java 17 installed')
```

## Step 3: Clone Repo (~10 sec)

```
%cd /content
!rm -rf DC
!git clone https://github.com/SONUVERMA11/DC.git
%cd /content/DC/flet_app
!ls -la
print('\n✅ Ready to build')
```

## Step 4: Build APK ⏳ (~5-10 min)

This uses `flet build apk` instead of Buildozer. Much faster and more reliable!

```
import os
os.environ['JAVA_HOME'] = '/usr/lib/jvm/java-17-openjdk-amd64'
os.environ['PATH'] = '/opt/flutter/bin:' + os.environ['PATH']

# Accept Android SDK licenses
!yes | flutter doctor --android-licenses 2>/dev/null || true
!flutter doctor

print('\n🔨 Building APK...\n')
!cd /content/DC/flet_app && flet build apk --verbose --project DC --org com.dc --product DC --build-version 2.0.0 --build-number 2
print('\n🏁 Build complete!')
```

## Step 5: Download APK

```
import glob, os
from google.colab import files

apk_files = glob.glob('/content/DC/flet_app/build/**/*.apk', recursive=True)
if not apk_files:
    import subprocess
    result = subprocess.run(['find', '/content/DC/flet_app', '-name', '*.apk', '-type', 'f'], capture_output=True, text=True)
    apk_files = [f.strip() for f in result.stdout.strip().split('\n') if f.strip()]

if apk_files:
    for apk in apk_files:
        size_mb = os.path.getsize(apk) / (1024 * 1024)
        print(f'✅ Found: {os.path.basename(apk)} ({size_mb:.1f} MB)')
        files.download(apk)
else:
    print('❌ No APK found. Check build output above.')
    !find /content/DC/flet_app -name '*.apk' -type f 2>/dev/null || echo 'None found'
```
